```python
import matplotlib.pyplot as plt

x = []  # Positions
y = []  # Coverage values

with open("/hpc/compgen/projects/fragclass/analysis/mvivekanandan/output/bam_outputs/chr1_L1-M13.donor.coverage.txt", "r") as coverage_file:
    for line in coverage_file:
        fields = line.strip().split("\t")
        x.append(int(fields[1]))  # Assuming the position is in the second column
        y.append(int(fields[2]))  # Assuming the coverage is in the third column

plt.plot(x, y)
plt.xlabel("Position")
plt.ylabel("Coverage")
plt.title("Coverage Plot")
plt.show()
```
